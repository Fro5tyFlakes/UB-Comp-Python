### Plotting Data in Python

Now we are going to learn how we can create plots of data in python. Plotting is very important in science, math, economics, marketing, politics ... The goal is to be able to display our data in the best way possible such that a reader can easily and quickly orient themselves across our plot. 


### Import MatPlotLib Package

```python
# import packages
import matplotlib.pyplot as plt # the plt is what we will use to call the library
%matplotlib inline  
# this makes the plots appear in the space below the cell
```

### Very basic syntax for plotting data

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25]) # First list is the x, second list is the y
plt.show()   # old versions need this 
```

```python
# using variables
x_vals = [i for i in range(0,100,15)]
y_vals = [i for i in range(100,400,45)]

# plot
plt.plot(x_vals,y_vals)
plt.show()
```

### Adding Title, axes labels, legend

```python
# plot
plt.plot(x_vals,y_vals, label = 'Data')  # add the label for the legend
plt.title('First Plot')  # Sets title to First Plot
plt.xlabel('X Values')   # x axis label
plt.ylabel('Y Values')   # y axis label
plt.legend() # creates legend.  [SIDE NOTE]: you can use plt.legend(loc='best') 
plt.show()                              # to move the legend in "best" location
```

### Plotting multiple data series on same figure. Setting color and marker style. 

```python
# data for scatter plot
x = [i for i in range(0,21,1)]
y = [0,2.5, 5.8, 11, 13.5, 28.9, 41.6, 45.1, 68.6, 84.2, 96.9, 125.4, 
     149.5, 166.69, 210.1, 233, 246.8, 295.2, 313.9, 379.9, 420]

# y = x^2 
y_model = []
for i in range(0,len(x),1):
    y_model.append(x[i]**2)
    
# plot both like this   first 2 inputs are the data for x,y axes
_ = plt.plot(x,y, 'b*', label = 'Collected Data')  # 3rd input is the marker color and style
_ = plt.plot(x,y_model, 'k--', label = 'Model') # 4th input is the label for the data series
plt.title('Multiple plots on same figure')
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.legend()
plt.show()
```

```python
# plot both like this   first 2 inputs are the data for x,y axes
_ = plt.plot(x,y, 'b*', label = 'Collected Data')  # 3rd input is the marker color and style
_ = plt.plot(x,y_model, 'k-', label = 'Model') # 4th input is the label for the data series
plt.title('Multiple plots on same figure')
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.ylim(30,250)   # how you control the zoom of the graph
plt.xlim(5,15)   # 1st input is starting point and 2nd input is the ending point
plt.legend()
plt.show()
```

### Other types
I will show how to make scatter plots, histograms, and bar graphs

```python
# scatter plot
# plot both like this   first 2 inputs are the data for x,y axes
plt.scatter(x,y, c = 'g',marker = '^',label = 'Data')  # 3rd input is the marker color and style

plt.title('Scatter Plot')
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.legend()
plt.show()
```

```python
# histogram
import random
x_new = [random.randint(0,10) for i in range(0,100,1)]

plt.hist(x_new, bins = 8 ,color = 'r', label = 'Collected Data')  
# the bins is the number of bars

plt.title('Histogram')
plt.xlabel('X Data')
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python
# bar graphs
people = ['B','C','D']
scores = [70, 83, 66, 97]

# plot
plt.barh(people,scores, label = 'Class 1') # x data, y data, label
plt.title('Bar Graph')
plt.ylabel('Person')
plt.xlabel('Test Score')
plt.legend()
plt.show ()
```
